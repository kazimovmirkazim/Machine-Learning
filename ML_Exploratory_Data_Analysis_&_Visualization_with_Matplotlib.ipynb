```python
c = np.arange (-20,20)
d = c ** 2
```

```python
plt.plot(c,d)
```

```python
plt.figure(figsize=(10,7))
plt.title('Parabola',size = 15,loc ='left')
plt.plot(c,d,'c--')
plt.grid('both')
plt.xlim([-10,15])
plt.ylim([0,150])
plt.xticks(np.arange(-10,15,2),rotation = 45)
plt.show()
```
