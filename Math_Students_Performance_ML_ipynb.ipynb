```python
X = df.drop('G3',axis=1)
y= df['G3'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.2,random_state=32)
```

```python

#  Linear Regression-dan istifadə səbəbləri:
# ✅ Sadə və başa düşülən — nəticələr asan interpretasiya olunur: "1 AZN artanda gəlir necə dəyişir?"
# ✅ Hesablama cəhətdən yüngül — böyük datalarda sürətlidir.
# ✅ Baseline model kimi istifadə olunur — ilk mərhələdə performansı ölçmək üçün.
# ✅ Feature əhəmiyyətini analiz etmək asandır — hansı dəyişən daha çox təsir edir görmək olur.
from sklearn.pipeline import Pipeline                      # Pipeline yaratmaq üçün
from sklearn.preprocessing import StandardScaler, OneHotEncoder  # Sayısal və kateqorik preprocessing üçün
from sklearn.compose import ColumnTransformer              # Sayısal və kateqorik sütunlar üçün ayrı-ayrı transformerlər
from sklearn.impute import SimpleImputer                   # NaN dəyərləri doldurmaq üçün
from sklearn.linear_model import LinearRegression          # Regression modeli
from sklearn.ensemble import RandomForestClassifier        # Klassifikasiya üçün Random Forest
from sklearn.metrics import mean_squared_error             # Regression üçün qiymətləndirmə metrikası
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error

num_pipeline = Pipeline  (
    [
      ('impute',SimpleImputer(strategy='median')),
      ('scale',StandardScaler())
    ]
)

cat_pipeline = Pipeline (
    [    ('impute',SimpleImputer(strategy='most_frequent')),
        ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
    ]

)
transformer = ColumnTransformer (
    [
        ('num',num_pipeline,num_features),
        ('cat',cat_pipeline,cat_features)
    ],remainder='passthrough'
)

estimator = LinearRegression()
full_pipeline = Pipeline (
    [
        ('transformer',transformer),
        ('estimotor', estimator)
    ]
)
```

```python
full_pipeline.fit(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test),full_pipeline.score(X_train,y_train)
```

```python
y_pred = full_pipeline.predict(X_test)
mse = mean_squared_error (y_test,y_pred)
mse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
r2
```

```python
from sklearn.linear_model import Ridge
estimator_1 = Ridge (alpha=0.1)
full_pipeline_1 = Pipeline (
    [
        ('transformer',transformer),
        ('estimator1',estimator_1)
    ]
)
```

```python
full_pipeline_1.fit(X_train,y_train)
```

```python
full_pipeline_1.score(X_test,y_test),full_pipeline_1.score(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import RidgeCV
from sklearn.svm import LinearSVR # Import LinearSVR from sklearn.svm


estimators = [
    ('lr', RidgeCV()),
    ('svr', LinearSVR(random_state=42, max_iter=10000)) # Increased max_iter to prevent convergence warning
]
reg = StackingRegressor(
    estimators=estimators,
    final_estimator=RandomForestRegressor(n_estimators=10,
                                          random_state=42)
)


reg.fit(full_pipeline.named_steps['transformer'].transform(X_train), y_train)
reg.score(full_pipeline.named_steps['transformer'].transform(X_test), y_test)
```
