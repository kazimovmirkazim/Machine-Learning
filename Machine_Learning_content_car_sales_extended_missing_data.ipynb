```python
X = df.drop ('Price',axis =1)
y = df['Price'].copy()
```

```python
df.dropna(subset='Price',inplace=True)
```

```python
df.isna().sum()
```

```python
# x- 80% x trein
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train, y_test = train_test_split (X,y,test_size =0.2) # bu kodla verilənlərin 80%-i təlim, 20%-i isə test üçün ayrılacaq.
```

```python
X_train
```

```python
y_train
```

```python
# w1 * Make + w2 * colour + w3 * Odometer + w4 * doors +b = y
```

```python
X_train
```

```python
# Encoder - Label Encoder,Ordinal encoder,one hot encoding
# one hot encoding - deyere gore column yaradir
X_train['Make'] = X_train['Make'].fillna('Missing')
X_test['Make'] = X_test['Make'].fillna('Missing')
X_train['Colour'] = X_train['Colour'].fillna('Missing')
X_test['Colour'] = X_test['Colour'].fillna('Missing')
X_train['Odometer (KM)'] = X_train['Odometer (KM)'].fillna(X_train['Odometer (KM)'].median())
X_test['Odometer (KM)'] = X_test['Odometer (KM)'].fillna(X_train['Odometer (KM)'].median())
X_train['Doors'] = X_train['Doors'].fillna(X_train['Doors'].median())
X_test['Doors'] = X_test['Doors'].fillna(X_train['Doors'].median())
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord_enc = OrdinalEncoder()
```

```python
X_train['Make'] = ord_enc.fit_transform(pd.DataFrame(X_train['Make']))
X_test['Make'] = ord_enc.transform(pd.DataFrame(X_test['Make']))
X_train['Colour'] = ord_enc.fit_transform(pd.DataFrame(X_train['Colour']))
X_test['Colour'] = ord_enc.transform(pd.DataFrame(X_test['Colour']))
```

```python
X_train.isna().sum()
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#regressor - x verilenleri ile y arasindaki elaqedir.
rfr = RandomForestRegressor()
```

```python
# fit - oyretmek demekdir gedir oyrenir
rfr.fit(X_train,y_train)
```

```python
rfr.score(X_train,y_train)
```

```python
rfr.score(X_test,y_test)
```
